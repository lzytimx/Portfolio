### 1.2 Merging datasets, converting columns to datetime objects, and simple feature engineering

```python
# Importing the monthly set into a (python native) list
cyclistic_data_folder = '/kaggle/input/cyclistic-dataset-2023/'
dataset_list = []
for i in os.listdir(cyclistic_data_folder):
    df = pd.read_csv(cyclistic_data_folder + i)
    dataset_list.append(df)
```

```python
# Next, I use the concat function from pandas to merge all the monthly dataset row-wise
cycle = pd.concat(dataset_list, axis=0)
# cycle.head()
```

```python
# returns the number of rows and columns in the dataset
cycle.shape
```

```python
# Creating some new columns. The most important being the duration of use for each ride (duration column)
cycle['started_at'] = pd.to_datetime(cycle['started_at'])
cycle['ended_at'] = pd.to_datetime(cycle['ended_at'])
cycle['duration'] = (cycle['ended_at'] - cycle['started_at']).dt.seconds/60
cycle['hour_start'] = cycle['started_at'].dt.hour
cycle['day_of_week_start'] = cycle['started_at'].dt.day_of_week
```

```python
cycle = cycle.sort_values(by='started_at')
cycle = cycle.reset_index(drop=True)
```

```python
# This line of code aims to understand what what are the outliers in the duration column of the dataset, 
# and how many records are exceed this duration.
ntile_95 = cycle['duration'].quantile(q=0.95)
# ntile_95
cycle[cycle['duration'] > ntile_95]['member_casual'].value_counts()
```

Question 4: What time of day do each of the customer archetypes usually loan bikes?

```python
pd.crosstab(index=cycle['member_casual'], columns=cycle['hour_start'], values=cycle['ride_id'], aggfunc='count', normalize='index')
```

```python
df_hour_start = pd.crosstab(index=cycle['member_casual'], columns=cycle['hour_start'], values=cycle['ride_id'], aggfunc='count', normalize='index')
# df_hour_start
num_of_casual = cycle['member_casual'].value_counts().iloc[0]
num_of_member = cycle['member_casual'].value_counts().iloc[1]

x = df_hour_start.columns.to_numpy()
y = df_hour_start.iloc[0, :].to_numpy().copy() #casual
y2 = df_hour_start.iloc[1, :].to_numpy().copy() # member

fig, ax = plt.subplots()
fig.set_size_inches(20, 12)

ax.plot(x, y, 'o--', label='casual')
ax.plot(x, y2, 'o-', label='member')

ax.set_xticks([i for i in range(0, 25, 6)])
ax.set_xticklabels([f'{i:02d}:00 hrs' for i in range(0, 25, 6)], fontsize=14)
ax.set_xlabel(xlabel='Time of Day', fontsize=16)
ax.set_xlim(-0.5, 23.5)

ax.set_yticks([i/100 for i in range(0, 11, 2)])
ax.set_yticklabels([f'{i}%' for i in range(0, 11, 2)], fontsize=14)
ax.set_ylim(-0.002, 0.11)

ax.set_title(label='Proportion of Rides Taken by Users at Different Times of Day', fontsize=24, pad=10, loc='left')

for xaxis, casual, member in zip(x, y, y2):
    casual_label =  f"{casual*100:.1f}%"
    member_label = f"{member*100:.1f}%"
    if casual > member:
        ax.annotate(casual_label, xy=(xaxis,casual), xycoords='data', xytext=(0, 15), textcoords='offset points', ha='center', fontsize=14, color='#E24A33')
        ax.annotate(member_label, xy=(xaxis,member), xycoords='data', xytext=(0, -15), textcoords='offset points', ha='center', fontsize=14, color='#348ABD')
    elif casual < member:
        ax.annotate(casual_label, xy=(xaxis,casual), xycoords='data', xytext=(0, -15), textcoords='offset points', ha='center', fontsize=14, color='#E24A33')
        ax.annotate(member_label, xy=(xaxis,member), xycoords='data', xytext=(0, 15), textcoords='offset points', ha='center', fontsize=14, color='#348ABD')

# Legend
ax.text(x=23.5, y=round(y[-1], 4), s="Casual Riders", color='#E24A33', fontsize=16, va='center')
ax.text(x=23.5, y=round(y[-1], 4)-0.004, s=f'{num_of_casual:,} rides', color='#E24A33', fontsize=16)
ax.text(x=23.5, y=round(y2[-1], 4), s="Members", color='#348ABD', fontsize=16, va='center')
ax.text(x=23.5, y=round(y2[-1], 4)-0.004, s=f'{num_of_member:,} rides', color='#348ABD', fontsize=16)


# Axes
ax.set_facecolor('None')
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')

# Specify Day/Night Cycle
ax.add_patch(Rectangle(xy=(-0.5,-0.002), width=6.5, height=0.12, facecolor='grey', alpha=0.2))
ax.add_patch(Rectangle(xy=(18,-0.002), width=6, height=0.12, facecolor='grey', alpha=0.2))

# fig.savefig('Rides_Time.png', transparent=False, dpi=80, bbox_inches="tight")
```
