```python
df_old_exclude_2022 = df_old[df_old['Year'] != 2022].copy()
```

```python
df_old_exclude_2022['Rank_Dense'] = df_old_exclude_2022.groupby('Year')[['Rank_Dense']].rank(method='dense')
df_old_exclude_2022['Rank_Competition'] = df_old_exclude_2022.groupby('Year')[['Rank_Dense']].rank(method='min')
```

```python
df_new_forcombining = df_new.copy()
```

```python
df_new_forcombining['Rank_Dense'] = df_new_forcombining.groupby('Year')[['Visa-free Destinations']].rank(ascending=False, method='dense')
df_new_forcombining['Rank_Competition'] = df_new_forcombining.groupby('Year')[['Visa-free Destinations']].rank(ascending=False, method='min')
```

```python
df_result = pd.concat([df_old_exclude_2022, df_new_forcombining])
```

```python
df_result = df_result.loc[:, ['Year', 'Country', 'Visa-free Destinations', 'Rank_Dense', 'Rank_Competition']]
```

```python
df_result = df_result.sort_values(['Year', 'Visa-free Destinations', 'Country'], ascending=False).reset_index(drop=True)
```

```python
df_result['Rank_Dense'] = df_result['Rank_Dense'].astype('int64')
df_result['Rank_Competition'] = df_result['Rank_Competition'].astype('int64')
```

```python
df_result['Country'] = df_result['Country'].replace('DR Congo', 'Congo (Dem. Rep.)')
df_result['Country'] = df_result['Country'].replace('Congo', 'Congo (Rep.)')
df_result['Country'] = df_result['Country'].replace('Taiwan', 'Taiwan (Chinese Taipei)')
df_result['Country'] = df_result['Country'].replace('Cape Verde', 'Cape Verde Islands')
df_result['Country'] = df_result['Country'].replace('Comoros', 'Comoro Islands')
df_result['Country'] = df_result['Country'].replace('Eswatini', 'eSwatini')
df_result['Country'] = df_result['Country'].replace('Hong Kong', 'Hong Kong (SAR China)')
df_result['Country'] = df_result['Country'].replace('Macau', 'Macao (SAR China)')
df_result['Country'] = df_result['Country'].replace('Palestine', 'Palestinian Territory')
df_result['Country'] = df_result['Country'].replace('Russia', 'Russian Federation')
df_result['Country'] = df_result['Country'].replace('São Tomé and Príncipe', 'Sao Tome and Principe')
df_result['Country'] = df_result['Country'].replace('Palau', 'Palau Islands')
df_result['Country'] = df_result['Country'].replace('Saint Kitts and Nevis', 'St. Kitts and Nevis')
df_result['Country'] = df_result['Country'].replace('Saint Lucia', 'St. Lucia')
df_result['Country'] = df_result['Country'].replace('Gambia', 'The Gambia')
df_result['Country'] = df_result['Country'].replace('Saint Vincent and the Grenadines', 'St. Vincent and the Grenadines')
df_result['Country'] = df_result['Country'].replace('East Timor', 'Timor-Leste')
df_result['Country'] = df_result['Country'].replace('Ivory Coast', "Cote d'Ivoire")
```
